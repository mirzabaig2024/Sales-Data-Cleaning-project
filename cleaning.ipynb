```python
import pandas as pd
import os
url = r"C:\Users\Lenovo\OneDrive\Desktop\Cleaning_practise\sample_sales_data_messy.xlsx"
Sales_table = pd.read_excel(url)
print(Sales_table.head(10))
```

```python
print(Sales_table.shape)
print(Sales_table.info())
print(Sales_table.describe())
```

```python
Sales_table = Sales_table.rename(columns={"Category":"Categories"})
Sales_table = Sales_table.rename(columns={"Region":"Regions"})
Sales_table
```

```python
dup_ordID = Sales_table["Order ID"].duplicated().sum()
null_values = Sales_table[Sales_table["Order ID"].isna()]
print(dup_ordID)
null_values
```

```python
seen = set()
Sales_table = Sales_table.dropna(subset=["Order ID"]) #droping null value in order id column
for i,order in Sales_table["Order ID"].items():
    
    if order in seen: #check weather it is in seen or not
        if i<len(Sales_table):
            if order+1 == Sales_table.loc[i+1,"Order ID"]: 
                Sales_table = Sales_table.drop(index = i)
            else:
                Sales_table.loc[i,"Order ID"] = order+1
   
    else:
        seen.add(order)
print("Successfully cleaned")


```

```python
Sales_table = Sales_table.reset_index(drop=True)
Sales_table
```

```python
dup = Sales_table.duplicated().sum()
print(f"duplicate in order id is {dup}")
```

```python
def format_date(value):
    value = str(value).strip()
    formats = [
        "%Y-%m-%d",   # 2024-05-24
        "%d-%m-%Y",   # 20-05-2024
        "%d/%m/%Y",   # 20/08/2023
        "%m/%d/%Y",   # in case some are US-style
        "%Y/%m/%d",
        "%B %d, %Y"
    ]
    for fmt in formats:
        try:
            return pd.to_datetime(value, format=fmt)
        except ValueError:
            continue
    return pd.NaT  # if nothing matches

Sales_table["Order Date"] = Sales_table["Order Date"].apply(format_date)

# Now format all into ONE consistent style
Sales_table["Order Date"] = Sales_table["Order Date"].dt.strftime("%d-%m-%Y")

print(Sales_table["Order Date"])
```

```python
null_dates = Sales_table["Order Date"].isnull().sum()
print(null_dates)
```

```python
Sales_table["Categories"] = Sales_table["Categories"].str.strip().replace({
    "Stationery": "Stationary", "furniture": "Furniture",
    "electronics": "Electronics", "office supplies": "Office Supplies"
})

Sales_table
```

```python
uni_items = Sales_table["Product Name"].unique()
print(uni_items)
```

```python
Sales_table["Product Name"] = Sales_table["Product Name"].str.title().replace("  "," ",regex=True)
uni_items = Sales_table["Product Name"].unique()
print(uni_items)
```

```python
Sales_table["Regions"] = Sales_table["Regions"].astype(str).str.strip()
```

```python
Sales_table["Sales Rep"] = Sales_table["Sales Rep"].str.strip()
```

```python
Sales_table["Regions"] = Sales_table["Regions"].str.capitalize()
uni_items = Sales_table["Regions"].unique()
print(uni_items)
```

```python
uni_items = Sales_table["Payment Method"].unique()
print(uni_items)
```

```python
Sales_table["Payment Method"] = Sales_table["Payment Method"].str.title()
uni_items = Sales_table["Payment Method"].unique()
print(uni_items)
```

```python
Sales_table["Quantity"] = Sales_table["Quantity"].replace("five",5)
leng = len(Sales_table["Quantity"])
for i in range(0,leng):
    if Sales_table.loc[i,"Quantity"] <= 0 or pd.isna(Sales_table.loc[i,"Quantity"]) :
        Sales_table = Sales_table.drop(index=i)
Sales_table
```

```python
Sales_table = Sales_table.reset_index()
Sales_table
```

```python
Sales_table["Unit Price"] = Sales_table["Unit Price"].astype(str).str.replace(r"[^\d.]","",regex=True)
Sales_table["Unit Price"] = pd.to_numeric(Sales_table["Unit Price"],errors= "coerce")
Sales_table
```

```python
null_value = Sales_table.isna().sum()
print(null_value)
```

```python
Sales_table["Sales Rep"] = Sales_table["Sales Rep"].fillna("Guest")
```

```python
Sales_table["Total Sales"] = Sales_table["Total Sales"].fillna(Sales_table["Quantity"]*Sales_table["Unit Price"])
```

```python
Sales_table["Payment Method"] = Sales_table["Payment Method"].fillna("Unknow")
```

```python
Sales_table["Sales Rep"] = Sales_table["Sales Rep"].str.replace("  "," ",regex=True) #removed double spaces from Salesrep column
```

```python
Sales_table["Customer Email"] = Sales_table["Sales Rep"].str.lower().replace(" ",".",regex=True) + "@email.com"
Sales_table
```

```python
#rearraning index values
Sales_table["index"] = range(0,len(Sales_table))
Sales_table
```

```python
ini_v = int(Sales_table.loc[0,"Order ID"])
print(type(ini_v))
Sales_table["Order ID"] = range(ini_v,ini_v+len(Sales_table))
Sales_table
```
